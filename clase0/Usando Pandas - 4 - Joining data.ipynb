```python
import pandas as pd
import numpy as np
```

```python
data_A = {'id': ['1', '2', '3', '4', '5'],
          'Atributo_1': ['A', 'C', 'E', 'G', 'I'],
          'Atributo_2': ['B', 'D', 'F', 'H', 'J']}
```

```python
data_B = {'id': ['1', '2', '6', '7', '8'],
          'Atributo_3': ['K', 'M', 'O', 'Q', 'S'],
          'Atributo_4': ['L', 'N', 'P', 'R', 'T']}
```

```python
data_C = {
        'id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'Atributo_3': [12, 13, 14, 15, 16, 17, 15, 12, 13, 23]}
```

```python
data_D = {'id': ['1', '2', '7', '4', '10'],
          'Atributo_1': ['A', 'F', 'A', 'G', 'A'],
          'Atributo_5': ['B', 'T', 'Z', 'K', 'L']}
```

```python
data_E = {'id': ['1', '2', '23', '14', '5', '8', '17'],
          'Atributo_1': ['A', 'A', 'A', 'G', 'G', 'B', 'D'],
          'Atributo_6': ['a', 'b', 'a', 'c', 'a', 'b', 'b']}
```

```python
data_F = {'id': ['8', '9', '10', '11', '12'],
          'Atributo_1': ['K', 'L', 'M', 'N', 'O'],
          'Atributo_2': ['P', 'Q', 'R', 'S', 'T']}
```

```python
data_G = {'id': ['8', '9', '10'],
          'Atributo_1': ['K', 'L', 'M'],
          'Atributo_2': ['P', 'Q', 'R']}
```

```python
df_A = pd.DataFrame(data_A)
df_B = pd.DataFrame(data_B)
df_C = pd.DataFrame(data_C)
df_D = pd.DataFrame(data_D)
df_E = pd.DataFrame(data_E)
df_F = pd.DataFrame(data_F)
df_G = pd.DataFrame(data_G)
```

## Right join

```python
df_A.merge(right=df_B, on="id", how="right")
```

## Outer join

```python
df_A.merge(right=df_B, on="id", how="outer")
```

## Anti join

```python
outer_join = df_A.merge(df_B, on="id", how='outer', indicator = True)
outer_join[~(outer_join._merge == 'both')].drop('_merge', axis=1)
```
