```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", kind="kde")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species", row="sex")
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.dtypes
```

```python
g = sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", hue="smoker", size="day")
g.ax.set_xlabel("Total Bill");
```

```python
penguins.head()
```

```python
g = sns.FacetGrid(data=penguins, col="sex", row="species")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", ax=g.axes[0,0])
sns.lineplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", ax=g.axes[0,1])
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", kind="hist")
```
