## Metricas

Entrenamos el modelo, pero para validar si está bien entrenado, debemos usar el dataset de testeo. 

Vamos a aplicar las siguientes metricas de evaluación usando scikit-learn:
- R2
- MAE
- MSE
- RMSE
- MAPE
- MPE

````python
#Primero obtenemos las predicciones del modelo
y_pred = regresion.predict(X_test)
````

````python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
````

````python
#sklean no tiene el error porcentual medio (MPE) lo vamos a crear nosotros
def mean_porcentual_error(yreal, ypred):

    return np.mean((yreal-ypred)/yreal)
````

````python
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)
mpe = mean_porcentual_error(y_test, y_pred)
print("R-cuadrado en test:", r2)
print("Error absoluto medio:", mae)
print("Error cuadratico medio:", mse)
print("Raiz de error cuadratico medio:", rmse)
print(f"Error absoluto porcentual medio: {mape*100:.2f}%")
print(f"Error porcentual medio: {mpe*100:.2f}%")
````

Estas no son las unicas métricas que se pueden calcular. Scikit-learn documenta varias [metricas de regresión](https://scikit-learn.org/stable/modules/model_evaluation.html#regression-metrics)

### Selección de modelo

Para seleccionar el modelo, vamos a usar una tecnica de eliminación hacia atrás y usando el criterio de información de Aikake (AIC) y criterio de información bayesiano (BIC). Como sklearn no nos da esas métricas para calcular, armamos una función que incorpora esto (esta función solo es valida para regresiones lineales).

Otra herramienta para hacer selección de modelos es utilizar una libreria especializada en herramientas estadisticas, llamada [statsmodels](https://www.statsmodels.org/stable/index.html) que automaticamente realizar diferentes test de hipotesis y calcula parámetros para ayudarnos a realizar este tipo de trabajo.

````python
def criterion(X, y, y_pred):
    # Agregamos uno porque hay que incorporar a la ordenada al origen
    d = X.shape[1]+1
    N = X.shape[0]

    # Calculamos los residuos al cuadrado
    residuals = y - y_pred
    Se = np.sum(residuals**2)
    # Se/N es la estimación de la varianza si los resuidos provienen
    # de una normal con media cero.

    # Calculamos la estimación del logaritmo de maxima similitud de la regresion lineal
    log_lik = np.log(2*np.pi) + np.log(Se/N) + 1
    log_lik *= -N/2

    #Calculamos ambos criterios
    aic = 2*d - 2*log_lik 
    bic = d*np.log(N) - 2*log_lik
   
    return aic, bic
````

Armamos el dataset para entrenar los modelos

````python
# Armamos array de Numpy con los features
X = dataset_with_dummies.loc[:,:"State_Florida"].values
y = dataset_with_dummies.loc[:,"Profit"].values

# Separando nuestro dataset en entrenamiento y testeo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

Las columnas de los arrays `X`, `X_train`, `X_test` son:
- 0: `'R&D Spend'`
- 1: `'Administration'`
- 2: `'Marketing Spend'`
- 3: `'State_California'`
- 4: `'State_Florida'`

Por lo que seleccionaremos los atributos en función de estos números. 

Armamos el modelo con todos los atributos y calculamos los criterios, para esto usasos siempre el set de entrenamiento:

````python
regresion = LinearRegression()
regresion.fit(X_train, y_train)
y_pred = regresion.predict(X_train)

aic, bic = criterion(X_train, y_train, y_pred)

print(f"AIC inicial es {np.round(aic)}")
print(f"BIC inicial es {np.round(bic)}")
````

Ahora creaos una funcion que dado una seleccion de columnas, entrena un modelo y obtiene AIC y BIC, esto nos va a facilitar ir avanzando paso a paso:

````python
def train_reg_model(X, y, columns):

    # Quitamos las columnas
    X_clear = X[:, columns].copy()
    
    model = LinearRegression()
    model.fit(X_clear, y)
    y_pred = model.predict(X_clear)

    return criterion(X_clear, y, y_pred)
````

Ahora creamos 5 modelos distintos, quitando para cada caso un atributo, y vemos el resultado de AIC y BIC

````python
aic, bic = train_reg_model(X_train, y_train, [1, 2, 3, 4])

print(f"Sacamos a {dataset_with_dummies.columns[0]}, el modelo nos da:")
print(f"AIC {np.round(aic, 1)}")
print(f"BIC {np.round(bic, 1)}")

aic, bic = train_reg_model(X_train, y_train, [0, 2, 3, 4])

print(f"Sacamos a {dataset_with_dummies.columns[1]}, el modelo nos da:")
print(f"AIC {np.round(aic, 1)}")
print(f"BIC {np.round(bic, 1)}")

aic, bic = train_reg_model(X_train, y_train, [0, 1, 3, 4])

print(f"Sacamos a {dataset_with_dummies.columns[2]}, el modelo nos da:")
print(f"AIC {np.round(aic, 1)}")
print(f"BIC {np.round(bic, 1)}")

aic, bic = train_reg_model(X_train, y_train, [0, 1, 2, 4])

print(f"Sacamos a {dataset_with_dummies.columns[3]}, el modelo nos da:")
print(f"AIC {np.round(aic, 1)}")
print(f"BIC {np.round(bic, 1)}")

aic, bic = train_reg_model(X_train, y_train, [0, 1, 2, 3])

print(f"Sacamos a {dataset_with_dummies.columns[4]}, el modelo nos da:")
print(f"AIC {np.round(aic, 1)}")
print(f"BIC {np.round(bic, 1)}")
````

Vemos que quitando `'State_California'` el modelo mejora más (tanto viendo AIC como BIC, recordar menor es mejor). Seguimos quitando columnas y viendo cuanto mejora el modelo:

````python
aic, bic = train_reg_model(X_train, y_train, [1, 2, 4])

print(f"Sacamos a {dataset_with_dummies.columns[0]}, el modelo nos da:")
print(f"AIC {np.round(aic, 1)}")
print(f"BIC {np.round(bic, 1)}")

aic, bic = train_reg_model(X_train, y_train, [0, 2, 4])

print(f"Sacamos a {dataset_with_dummies.columns[1]}, el modelo nos da:")
print(f"AIC {np.round(aic, 1)}")
print(f"BIC {np.round(bic, 1)}")

aic, bic = train_reg_model(X_train, y_train, [0, 1, 4])

print(f"Sacamos a {dataset_with_dummies.columns[2]}, el modelo nos da:")
print(f"AIC {np.round(aic, 1)}")
print(f"BIC {np.round(bic, 1)}")

aic, bic = train_reg_model(X_train, y_train, [0, 1, 2])

print(f"Sacamos a {dataset_with_dummies.columns[4]}, el modelo nos da:")
print(f"AIC {np.round(aic, 1)}")
print(f"BIC {np.round(bic, 1)}")
````

## Bias-Variance

Veamos como varia el error MSE dados el efecto de la varianza y el sesgo. Para ello, vamos a usar datos sintéticos.

$$y = f(x) = 0.00004x^5 - 0.05 x^4 - 0.04 x^3 + 0.1 x$$

````python
# Creemos un dataset de 100 muestras entre 0 y 10
X = np.random.rand(1000) * 10

# Apliquemos la formula
# Pero el target que vamos a usar para entrenar le agregemos ruido gaussiano
y = 0.00004*X**5 - 0.05 * X**4 - 0.04*X**3 + 0.1*X + np.random.randn(1000)*50

X_real = np.linspace(0, 10, 100, endpoint=True)
y_real = 0.00004*X_real**5 - 0.05 * X_real**4 - 0.04*X_real**3 + 0.1*X_real
````

````python
plt.figure(figsize=(7, 5))
plt.scatter(X, y, color = "g")
plt.plot(X_real, y_real, color = "black", label="Relación real")
plt.legend()
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14);
````

Ahora vamos a crear 10 modelos polinomiales, empezando por el lineal hasta el de grado 5, para cada caso se va a generar diferentes datasets de 100 muestras

````python
# Aplicamos este reshape para poder entrenar
X = X.reshape(-1, 1)
X_real = X_real.reshape(-1, 1)

models_values = list()
mse_array = np.zeros([10, 100])

for order in range(1, 11):
    for tt in range(100):
        # Obtenemos 150 muestras al azar de la población
        perm = np.random.permutation(1000)[:150]

        # Entrenamos con 100 muestras
        X_train = X[perm[:30]].copy()
        y_train = y[perm[:30]].copy()

        # Y testeamos con 50
        X_test = X[perm[100:120]].copy()
        y_test = y[perm[100:120]].copy()

        # Creamos el modelo
        if order != 1:
            pol_reg = PolynomialFeatures(degree = order)
            X_poly = pol_reg.fit_transform(X_train)
            X_poly_test = pol_reg.transform(X_test)
        else:
            X_poly = X_train
            X_poly_test = X_test

        # Entrenamos
        model = LinearRegression()
        model.fit(X_poly, y_train)

        # Predecimos con el datset de testeo
        y_pred = model.predict(X_poly_test)

        # Calculamos el error y lo guardamos
        mse_array[order-1, tt] = mean_squared_error(y_test, y_pred)

        # Guardamos una prediccion para graficar
        if tt == 0:
            if order != 1:
                X_poly_real = pol_reg.transform(X_real)
            else: 
                X_poly_real = X_real
            
            models_values.append(model.predict(X_poly_real))

# Calculamos el MSE promedio para cada orden de polinomio
mse_mean = np.mean(mse_array, axis=1)
mse_std = np.std(mse_array, axis=1)
````

````python
X = X.reshape(-1)
X_real = X_real.reshape(-1)

plt.figure(figsize=(7, 5))

plt.scatter(X, y, color = "#505050", alpha=0.2, zorder=0)

plt.plot(X_real, y_real, color = "black", label="Relación real", zorder=10)

for index, y_plot in enumerate(models_values):
    plt.plot(X_real,y_plot, label=f"Order {index+1}", zorder=10)

plt.legend()
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14);
````

````python
X = X.reshape(-1)
X_real = X_real.reshape(-1)

plt.figure(figsize=(7, 5))

plt.errorbar(np.arange(1, 11), mse_mean, yerr=mse_std)
plt.yscale("log") 
plt.xlabel("Orden polinomio", fontsize=14)
plt.ylabel("MSE promedio", fontsize=14);
````

Vemos como el error empieza en valores altos y va disminuyendo a medida que el orden aumenta, cerca de la zona de orden 4 y luego empieza a aumentar. Como es datos sintéticos podemos saber que el error alto de MSE cuando el orden es bajo, la fuente de error más alta es el error de bias, en cambio a medida que aumenta, el error es debido a la varianza. Esto es notable también viendo el desvío estándar entre cálculos para un mismo orden del MSE.

## Regresión Lasso y Ridge

Para ver el efecto de las penalizaciones L1 y L2 en la regresión lineal, vamos a ver el dataset de las 50 startups. Ya vimos, cuando hicimos Selección de modelo mediante eliminación hacia atrás, encontramos que con solo dos atributos (`"R&D Spend"`, `"Marketing Spend"`), obtuvimos un mejor modelo. Veamos cómo nos afecta el hecho de usar estas penalidades.

````python
#Cargamos el dataset y aplicamos las transformaciones que ya vimos previamente.
dataset = pd.read_csv("datasets/50_Startups.csv") 
dataset.head() 
dataset_with_dummies = pd.get_dummies(data=dataset, columns=['State'])
dataset_with_dummies.drop(columns="State_New York", inplace=True)
dataset_with_dummies = dataset_with_dummies.astype('float')

dataset_with_dummies = dataset_with_dummies[["R&D Spend", "Administration", "Marketing Spend", "State_California", "State_Florida", "Profit"]]
dataset_with_dummies.head()
````

````python
# Armamos array de Numpy con los features
X = dataset_with_dummies.loc[:,:"State_Florida"].values
# Y con la variable dependendiente (target)
y = dataset_with_dummies.loc[:,"Profit"].values
````

````python
# Separando nuestro dataset en entrenamiento y testeo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Datos de train y de test
print("Dimension de X_train:",X_train.shape)
print("Valores de y_train:",y_train.size)
print("Dimension de X_test:",X_test.shape)
print("Valores de y_test:",len(y_test))
````

Para este tipo de regresiones, debemos aplicar [normalización de variables](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html).

````python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()

X_train = sc_X.fit_transform(X_train) # usamos fit para crear el escalado y aplicar conjunto de entrenamiento
X_test = sc_X.transform(X_test) # solo hacemos el transform ya que fue escalado anteriormente
````

````python
# Primero aplicamos la regresión lineal con todas las variables y el mejor modelo que vimos
# eliminacion hacia adelante
regresion = LinearRegression()
reg_stepwise = LinearRegression()

# Entrenamos el modelo
regresion.fit(X_train, y_train)
reg_stepwise.fit(X_train[:,[0,2]], y_train)

# Realizamos la parte de testeo
y_pred = regresion.predict(X_test)

r2_linear = r2_score(y_test, y_pred)
mse_linear = mean_squared_error(y_test, y_pred)
rmse_linear = np.sqrt(mse_linear)

y_pred_stepwise = reg_stepwise.predict(X_test[:,[0,2]])

r2_stepwise = r2_score(y_test, y_pred_stepwise)
mse_stepwise = mean_squared_error(y_test, y_pred_stepwise)
rmse_stepwise = np.sqrt(mse_stepwise)
````

En sklearn tenemos a las dos regresiones, [Ridge](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html) y [Lasso](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Lasso.html):

````python
from sklearn.linear_model import Ridge, Lasso
````

Grafiquemos los valores de cada coeficiente a medida que aumenta $\lambda$ para ambas regresiones

````python
coeffs_lasso = np.zeros([5000, 5])
coeffs_ridge = np.zeros([5000, 5])

# Guardamos los coeficientes de la regresion lineal (lambda = 0)
coeffs_lasso[0, :] = regresion.coef_.copy()
coeffs_ridge[0, :] = regresion.coef_.copy()
lamb_array = np.arange(0, 10000, 2)

# Calculamos los coeficientes para diferentes valores de lambda
for index, lamb in enumerate(lamb_array):

    if index == 0:
        continue

    # Creamos los modelos
    lasso_model = Lasso(alpha=lamb)
    ridge_model = Ridge(alpha=lamb)

    # Los entrenamos
    lasso_model.fit(X_train, y_train)
    ridge_model.fit(X_train, y_train)

    # Guardamos los coeficientes de la regresiones
    coeffs_lasso[index, :] = lasso_model.coef_.copy()
    coeffs_ridge[index, :] = ridge_model.coef_.copy()
````

````python
f, ax = plt.subplots(1, 2, figsize=(9, 3.5))

for i in range(coeffs_lasso.shape[1]):
    ax[0].plot(lamb_array, coeffs_lasso[:, i])

    ax[1].plot(lamb_array, coeffs_ridge[:, i], label=dataset_with_dummies.columns[i])

ax[1].legend()
ax[0].set_xlim([-100, 8000])
ax[1].set_xlim([-100, 4000])

ax[0].set_ylabel("Coefficients")
ax[0].set_xlabel("$\lambda$")
ax[1].set_xlabel("$\lambda$")

ax[0].set_title("Lasso (L1)")
ax[1].set_title("Ridge (L2)")

plt.tight_layout();
````

Vemos los efectos de las regresiones, donde en la regresión de Ridge tiende a cero todos los coeficientes, mientras que en la de Lasso, los diferentes coeficientes se van haciendo cero, en el mismo orden de cuando hicimos la selección mediante eliminación hacia atrás.

Veamos un ejemplo de dos modelos de cada uno. OBS: la elección de los valores de lambda fue hecha con un método que veremos más adelante (GridSearch).

````python
# Creamos los modelos
lasso_model = Lasso(alpha=871.45)
ridge_model = Ridge(alpha=1)

# Entrenamos los modelos
lasso_model.fit(X_train, y_train)
ridge_model.fit(X_train, y_train)

# Realizamos la parte de testeo
y_pred_lasso = lasso_model.predict(X_test)
y_pred_ridge = ridge_model.predict(X_test)

r2_lasso = r2_score(y_test, y_pred_lasso)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
rmse_lasso = np.sqrt(mse_lasso)

r2_ridge = r2_score(y_test, y_pred_ridge)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
rmse_ridge = np.sqrt(mse_ridge)

print("Modelo de regresión lineal")
print("R-cuadrado en test:", r2_linear)
print("Raiz de error cuadratico medio:", rmse_linear)

print("\nModelo de regresión lineal usando eliminación hacia atrás")
print("R-cuadrado en test:", r2_stepwise)
print("Raiz de error cuadratico medio:", rmse_stepwise)

print("\nLasso")
print("R-cuadrado en test:", r2_lasso)
print("Raiz de error cuadratico medio:", rmse_lasso)

print("\nRidge")
print("R-cuadrado en test:", r2_ridge)
print("Raiz de error cuadratico medio:", rmse_ridge)
````

Vemos que la regresión de Lasso nos dio mejor resultado, el cual es parecido al resultado obtenido en la selección de modelo mediante eliminación hacia atrás. Es un poco mejor, pero claramente muestra una mejora con respecto a una regresión lineal de todos los atributos.

Veamos el coeficiente del modelo de regresión Lasso, 

````python
print(f"Los coeficientes de la regresión Lasso con lambda=871 son {lasso_model.coef_}")
````

Vemos que los coeficientes de la ubicación de las startups son cero. Por lo que, este modelo no tiene en cuenta esos dos valores, por lo que se puede usar el modelo sin pasar esos atributos. 

La regresión Lasso no puede encontrar el mejor modelo de solo usar los dos atributos `"R&D Spend"` y `"Marketing Spend"` porque la penalización para que la columna `"Administration"` sea cero le afecta mucho también a los valores de las columnas `"R&D Spend"` y `"Marketing Spend"`, y por consiguiente modelos Lasso sin el atributo `"Administration"` performan peor que la regresión lineal que obtuvimos mediante selección de modelo.

Por otro lado, la regresión de Ridge no mejora el resultado obtenido. Esto se debe a que el problema que tenemos, la regresión lineal no presenta problemas de error de varianza. Ridge tiene sentido cuando tenemos muchos atributos y pocas observaciones.
